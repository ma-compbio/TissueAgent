```python
import importlib
import os
import sys

from IPython.display import Image, display
from langchain_core.messages import HumanMessage
from pydantic import SecretStr
from queue import Queue

sys.path.insert(0, "/Users/dustinm/projects/research/ma-lab/SpatialAgent/src")

gpt_model = "gpt-4o"
api_key = SecretStr(os.getenv("OPENAI_API_KEY", ""))
```

```python
import graph.graph as graph

state_queue = Queue()
app = graph.create_spatialagent_graph(gpt_model, api_key, state_queue).compile()

try:
    display(Image(app.get_graph().draw_mermaid_png()))
except Exception:
    print("Image display failed")
```

```python
prompt = HumanMessage(
    content="Hello! What are some things you can do?"
)

final_state = app.invoke({"messages": [prompt]})

print(final_state["messages"][-1].content)
```
