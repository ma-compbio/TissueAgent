```python
import inspect, re
from pathlib import Path

docs = Path("docs")
docs.mkdir(parents=True, exist_ok=True)

def safe_name(obj):
    mod = getattr(obj, "__module__", "")
    qual = getattr(obj, "__qualname__", getattr(obj, "__name__", obj.__class__.__name__))
    return re.sub(r"[^\w.-]", "_", f"{mod}.{qual}")

for obj in methods:
    try:
        sig = str(inspect.signature(obj))
    except Exception:
        sig = "()"
    qual = getattr(obj, "__qualname__", getattr(obj, "__name__", obj.__class__.__name__))
    doc = inspect.getdoc(obj) or ""
    s = f"{qual}{sig}\n\n{doc}"
    (docs / f"{safe_name(obj)}.txt").write_text(s)
```
